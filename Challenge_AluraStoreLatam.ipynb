# 1. Ingresos Totales por Tienda.

En este item se tomaron todos los ingresos totales de cada tienda desde el 01 de enero 2020 hasta el 30 de marzo 2023, esto debido a que la tienda cuatro no presenta datos para el 31 de marzo , como las otras tres tiendas.

Los resultados se presentan a continuacion:



```python
import pandas as pd

tienda1 = pd.read_csv(url, parse_dates=['Fecha de Compra'], dayfirst=True)
tienda2 = pd.read_csv(url2, parse_dates=['Fecha de Compra'], dayfirst=True)
tienda3 = pd.read_csv(url3, parse_dates=['Fecha de Compra'], dayfirst=True)
tienda4 = pd.read_csv(url4, parse_dates=['Fecha de Compra'],dayfirst=True)

inicio = pd.to_datetime('01/01/2020', dayfirst=True)
fin = pd.to_datetime('30/03/2023', dayfirst=True)

tienda1_filtrada = tienda1[(tienda1['Fecha de Compra'] >= inicio) & (tienda1['Fecha de Compra'] <= fin)]
tienda2_filtrada = tienda2[(tienda2['Fecha de Compra'] >= inicio) & (tienda2['Fecha de Compra'] <= fin)]
tienda3_filtrada = tienda3[(tienda3['Fecha de Compra'] >= inicio) & (tienda3['Fecha de Compra'] <= fin)]
tienda4_filtrada = tienda4[(tienda4['Fecha de Compra'] >= inicio) & (tienda4['Fecha de Compra'] <= fin)]

ingreso_tienda1 = tienda1_filtrada['Precio'].sum()
ingreso_tienda2 = tienda2_filtrada['Precio'].sum()
ingreso_tienda3 = tienda3_filtrada['Precio'].sum()
ingreso_tienda4 = tienda4_filtrada['Precio'].sum()

print(f'Ingreso total de la tienda 1 (2020-2023): {ingreso_tienda1}')
print(f'Ingreso total de la tienda 2 (2020-2023): {ingreso_tienda2}')
print(f'Ingreso total de la tienda 3 (2020-2023): {ingreso_tienda3}')
print(f'Ingreso total de la tienda 4 (2020-2023): {ingreso_tienda4}')
```

La visualizacion de los ingresos totales por tienda, se muestran en el siguiente grafico de barras:

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Datos de ingresos (puedes usar los que ya calculaste)
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

colores = ['blue', 'green', 'yellow', 'crimson']

# Crear gráfico
plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos, color=colores, width=0.4, align='center')

plt.ylim(1000000000, 1200000000)

plt.title('Ingresos Totales por Tienda (Ene 2020 - Mar 2023)')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total ($)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, ingreso in enumerate(ingresos):
    texto = f"{ingreso:,.0f}".replace(",", "X").replace(".", ",").replace("X", ".")
    plt.text(i, ingreso + max(ingresos)*0.01, texto, ha='center', va='top', color='black', fontsize=8)

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.tight_layout()
plt.show()
```

ANALISIS

Se aprecia clamente que para el periodo del 01 de enero 2020 al 30 de marzo 2023, la tienda 4 es la que presenta menor recaudación.

Cabe señalar, que para un analisis mas exhaustivo de los ingresos de cada tienda, habria sido necesario conocer el costo total por articulo vendido, antecedente que la base de datos de las cuatro tiendas no entrega.
En consecuencia, y para un mejor analisis no es posible obtener el margen de ganancia por articulo vendido, dato que podria habernos servido en la recomendación final.

# 4. Productos más y menos vendidos

En este item, se analizan los articulos que fueron más vendidos y menos vendidos para una de las tiendas.

```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

destacados = []

for tienda_df, nombre in zip(tiendas, nombres):
    ventas = tienda_df['Producto'].value_counts()
    mas_vendido = ventas.idxmax()
    menos_vendido = ventas.idxmin()

    destacados.append({
        'Tienda': nombre,
        'Producto Más Vendido': mas_vendido,
        'Cantidad Más Vendida': ventas.max(),
        'Producto Menos Vendido': menos_vendido,
        'Cantidad Menos Vendida': ventas.min()
    })

df_destacados = pd.DataFrame(destacados)

print("Productos más y menos vendidos por tienda:\n")
print(df_destacados)
```

Por ejemplo, en la tienda 1  su mayor venta fue microondas y tuvo menos ventas en los auriculares con microfono. Así, la tienda 4 vendió mas camas box y tuvo menos cantidad de guitarras electricas vendidas.

```python
import pandas as pd

# URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Carga de datos
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Lista para resultados
destacados = []

# Procesamiento por tienda
for tienda_df, nombre in zip(tiendas, nombres):
    ventas = tienda_df['Producto'].value_counts()

    top_5 = ventas.head(5)
    bottom_5 = ventas.tail(5)

    for producto, cantidad in top_5.items():
        destacados.append({
            'Tienda': nombre,
            'Tipo': 'Más Vendido',
            'Producto': producto,
            'Cantidad': cantidad
        })

    for producto, cantidad in bottom_5.items():
        destacados.append({
            'Tienda': nombre,
            'Tipo': 'Menos Vendido',
            'Producto': producto,
            'Cantidad': cantidad
        })

# Crear DataFrame final
df_destacados = pd.DataFrame(destacados)

# Mostrar resultados
print("Top 5 productos más y menos vendidos por tienda:\n")
print(df_destacados)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

tiendas_unicas = df_destacados['Tienda'].unique()

for tienda in tiendas_unicas:
    plt.figure(figsize=(6, 3))

    datos_tienda = df_destacados[df_destacados['Tienda'] == tienda]
    datos_tienda = datos_tienda.sort_values(by='Cantidad', ascending=False)

    ax = sns.barplot(
        data=datos_tienda,
        x='Cantidad',
        y='Producto',
        hue='Tipo',
        palette={'Más Vendido': 'blue', 'Menos Vendido': 'yellow'},
        dodge=True,
        width=0.4
    )

    plt.title(f"Productos Más y Menos Vendidos - {tienda}", fontsize=10)
    plt.xlabel("Cantidad Vendida", fontsize=9)
    plt.ylabel("Producto")
    plt.xticks(fontsize=9)
    plt.yticks(fontsize=9)
    plt.xlim(20, 70)  # De 0 a 50, ajusta según tus datos

    plt.legend(title='Tipo de Venta',fontsize=8,title_fontsize=9)

    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f', fontsize=8, padding=3)

    plt.tight_layout()
    plt.show()
```

El costo de envio se observa en el siguiente grafico de torta.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de costos de envío promedio
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos_envio = [26018.61, 25216.24, 24805.68, 23459.46]

# Función para mostrar valor real y porcentaje  en el gráfico
def mostrar_valor_y_porcentaje(pct, allvals):
    total = sum(allvals)
    valor_absoluto = int(round(pct * total / 100.0))
    return f"${valor_absoluto:,.0f}\n({pct:.1f}%)"

colores_personalizados = ['#1f77b4', '#2ca02c', '#FFFF00', '#d62728']  # Azul, Verde, Amarillo, Rojo
plt.figure(figsize=(5, 5))
plt.pie(costos_envio, labels=tiendas, colors=colores_personalizados,
        autopct=lambda pct: mostrar_valor_y_porcentaje(pct, costos_envio),
        textprops={'fontsize': 10, 'fontfamily': 'sans-serif'})
plt.title('Distribución del Costo de Envío Promedio por Tienda', fontname='sans-serif', fontsize=10)
plt.show()
```

ANALISIS

Analizando el costo de envio por cada tienda , encontramos que la tienda 4 presente el menos costo de envio con un valor de 23.459 y la tienda 1  con el costo mas alto de 26.019.

Los datos ni la base de datos indica si este costo es del cliente o de las tiendas. Por cuanto, no se puede descontar este costo de los ingresos de las tiendas ni tampoco indicar que es mejor comprar en la tienda 4 porque este costo es menor.

Tal vez la tienda 1 presenta un costo promedio mas elevado debido a que tiene un mayor volumen de ventas que las otras tiendas, y por el contrario la tienda 4 al tener menos ingresos  y ventas presenta un costo menor en los envios
